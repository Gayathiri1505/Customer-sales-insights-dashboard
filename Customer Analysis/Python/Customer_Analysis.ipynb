```python
obj_graph=df.select_dtypes(include=['object'])
obj_graph
```

```python
for i in obj_graph.columns:
    column_mode = df[i].mode()[0]
    df[i] = df[i].fillna(column_mode)
```

```python
df.isnull().sum()
```

```python
fgraph=df.select_dtypes(include=['float'])
fgraph
```

```python
fgraph.hist(color='g')
```

```python
df[fgraph.columns] = df[fgraph.columns].fillna(df[fgraph.columns].median())
```

```python
df.isnull().sum()
```

```python
df = df.drop_duplicates()
df
```
